## Make into DataFrame

```python
pd.DataFrame(response.json()['results']).head(3)
```

```python
new_df = pd.DataFrame()
```

```python
for i in range(1,500):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=b8d0f958de8242ac18a57db0c313c5c1&language=en-US&page={}'.format(i))
    
    df = pd.DataFrame(response.json()['results'])
    
    ## use concat
    new_df = new_df.append(df,ignore_index=True)
    
    
    
```
